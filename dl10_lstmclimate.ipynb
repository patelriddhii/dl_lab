```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense


url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url)

df.head()


city_data = {
    "Delhi": df[['Temp']].copy(),
    "Mumbai": df[['Temp']].copy() + 2,   # slightly warmer
    "Pune": df[['Temp']].copy() - 1      # slightly cooler
}


def prepare_data(dataset, time_step=30):
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(dataset)

    X, y = [], []

    for i in range(time_step, len(scaled_data)):
        X.append(scaled_data[i-time_step:i, 0])
        y.append(scaled_data[i, 0])

    X, y = np.array(X), np.array(y)
    X = X.reshape(X.shape[0], X.shape[1], 1)

    return X, y, scaler


models = {}
scalers = {}
test_sets = {}

for city in city_data:
    print(f"\nTraining for {city}...")

    dataset = city_data[city].values

    X, y, scaler = prepare_data(dataset)

    split = int(len(X)*0.8)

    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1],1)))
    model.add(LSTM(50))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')

    model.fit(X_train, y_train, epochs=5, batch_size=32, verbose=1)

    models[city] = model
    scalers[city] = scaler
    test_sets[city] = (X_test, y_test)


for city in models:
    print(f"\nResults for {city}")

    model = models[city]
    scaler = scalers[city]
    X_test, y_test = test_sets[city]

    predictions = model.predict(X_test)

    predictions = scaler.inverse_transform(predictions)
    actual = scaler.inverse_transform(y_test.reshape(-1,1))

    plt.plot(actual, label='Actual')
    plt.plot(predictions, label='Predicted')
    plt.title(f"{city} Temperature Prediction")
    plt.legend()
    plt.show()


for city in models:
    model = models[city]
    scaler = scalers[city]
    X_test, y_test = test_sets[city]

    predictions = model.predict(X_test)

    predictions = scaler.inverse_transform(predictions)
    actual = scaler.inverse_transform(y_test.reshape(-1,1))

    rmse = np.sqrt(mean_squared_error(actual, predictions))

    print(f"{city} RMSE:", rmse)


for city in models:
    print(f"\nFuture prediction for {city}")

    model = models[city]
    scaler = scalers[city]

    dataset = city_data[city].values
    scaled_data = scaler.transform(dataset)

    last_seq = scaled_data[-30:]
    current = last_seq.reshape(1,30,1)

    future = []

    for _ in range(7):
        pred = model.predict(current)[0][0]
        future.append(pred)

        current = np.append(current[:,1:,:], [[[pred]]], axis=1)

    future = scaler.inverse_transform(np.array(future).reshape(-1,1))

    print("Next 7 Days:", future)
```
